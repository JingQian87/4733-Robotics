```
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```
def dphidt(y,t):
    dydt = np.sin(t)
    return dydt
  
def dthetadt(y, t, phi):
    dydt = np.tan(phi)
    return dydt

def dxdt(x, t, theta):
  dydt = np.cos(theta)
  return dydt

def dydt(y, t, theta):
  dydt = np.sin(theta)
  return dydt
  
# initial condition
x, y, theta, phi = 0, 0, 0, 0
xs, ys = [0], [0]
phis=[phi]
thetas =[0]

# time points
t = np.linspace(0,20,21)

for i in range(20):
  phi = (odeint(dphidt,phi,[i, i+1]))[-1]
  phis.append(phi)
  theta = (odeint(dthetadt, theta, [i, i+1], args = (phi,)))[-1]
  thetas.append(theta)
  x = (odeint(dxdt, x, [i, i+1], args = (theta,)))[-1]
  xs.append(x)
  y = (odeint(dydt, y, [i, i+1], args = (theta,)))[-1]
  ys.append(y)

# solve ODE
#phi = odeint(model, 0, t)
#theta = odeint(model,theta,t)

plt.plot(t, xs)
#plt.plot(t,phis)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.show()
```
